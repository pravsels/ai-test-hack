```python
import pandas as pd
import inflect
ie = inflect.engine()
```

```python
word_list = [
             'england', 'france', 'germany', 'russia', 'austria', 'america', 'canada', 'china', 'india', 'brazil',
             'dune', 'mars', 'jupiter', 'saturn', 'venus', 'mercury', 'uranus', 'giedi', 'caladan', 'kaitan', 'salusa',
             'alexander', 'caesar', 'napoleon', 'genghis', 'cyrus', 'tamerlane', 'attila', 'stalin',
             'musk', 'bezos', 'zuckerberg', 'altman', 'jobs', 'gates', 'hughes', 'page', 'brin', 'ellison', 
             'edison', 'wright', 'galton', 'darwin', 'turing', 'newton', 'tesla', 'faraday', 'watson', 'crick',
             'rowling', 'christie', 'king', 'tolstoy', 'austen', 'tolkien', 'orwell', 'dickens', 'dostoevsky', 'gaiman',
             'locke', 'kant', 'hume', 'nietzsche', 'machiavelli', 'rand', 'rousseau', 'hobbes', 'mill', 'popper', 'voltaire',
             'euler', 'gauss', 'ramanujan', 'neumann', 'hilbert', 'fourier', 'fibonacci', 'laplace', 'bernoulli', 'poisson',
             
            ]
```

```python
prompts = []
answers = []
```

```python
for word in word_list:
  prompt = "Q: Count the number of letters in the word '" + word + "' ? \n A: "
  answer = len(word)
  answer = str(answer) + ',' + ie.number_to_words(answer)
  prompts.append(prompt)
  answers.append(answer)
  for letter in set(word): 
    prompt = "Q: How many occurences of the letter '" + letter + "' occur in the word '" + word + "' ? \n A: "
    answer = word.count(letter)
    answer = str(answer) + ',' + ie.number_to_words(answer)
    prompts.append(prompt)
    answers.append(answer)
```

```python
df = pd.DataFrame({'prompt': prompts, 'answer': answers})
```
